```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea

print("Imported Successfully")
```

                                                            Loading & Cleaning Dataset

```python
# Loading Data to the notebook
data = pd.read_csv("Dataset.csv")

# Standardize column names (remove spaces, lowercase)
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_")

# Remove fully duplicate rows
data = data.drop_duplicates()
```

```python
# size of our data
data.size

# shape of data
data.shape
```

```python
# columns present in dataset
print(data.columns)
```

                                                               Level 1 -Task 1

```python
                                   # Question 1 - Determine the top three most common cuisines in the dataset.

# First we need to split cuisine elements & to find out top 3 will count how many elements each cuisine have
data['cuisines'] = data['cuisines']
cnt = data['cuisines'].str.split(',').explode().str.strip().value_counts()
print(cnt)

top_3 = cnt.head(3)
print('The Top 3 most common cuisines:', top_3)

                                                    # --- Creating a Bar Chart ---
plt.figure(figsize=(6, 3)) # Set figure size

# Create a barplot using Seaborn
plt.bar(top_3.index, top_3.values, color=['purple', 'orange', 'green'])

# Adding labels and title
plt.xlabel('Cuisines', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.title('Top 3 Most Common Cuisines', fontsize=14)

# Display the chart
plt.show()

# horizontal bar graph presentation
plt.figure(figsize=(6, 4))
sea.barplot(y=top_3.index, x=top_3.values, palette='cool', orient='h')
plt.ylabel('Cuisines', fontsize=12)
plt.xlabel('Number of Restaurants', fontsize=12)
plt.title('Top 3 Cuisines (Horizontal Bar Chart)', fontsize=14)

plt.show()
```

```python
# Total number of restaurants (row count of restaurant id)
total_restaurants = len(data)

# Calculate the percentage of restaurants serving each top cuisine
percentage_top_cuisines = round(cnt / total_restaurants * 100, 2)

# Display only the top 3 cuisines' percentages
top_3_percentages = percentage_top_cuisines.head(3)

print('Percentage of restaurants serving the top cuisines:')
print(top_3_percentages)

# Create a bar chart
Top_cuisines_percentage = {'North Indian': 41.46,
                           'Chinese': 28.64,
                           'Fast Food': 20.79}

plt.figure(figsize=(6, 3))  # Set figure size

# Create a bar chart for percentages
plt.bar(Top_cuisines_percentage.keys(), Top_cuisines_percentage.values(), color=['blue', 'orange', 'green'])

# Adding labels and title
plt.xlabel('Cuisines', fontsize=12)
plt.ylabel('Percentage of Restaurants (%)', fontsize=12)
plt.title('Percentage of Restaurants Serving Top Cuisines', fontsize=14)

# Display the chart
plt.show()
```
