```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea

print("Imported Successfully")
```

```python
# Loading data from dataset
data = pd.read_csv('Dataset.csv')

# columns present in dataset
print(data.columns)
```

```python
# Question 1 - Create a histogram or bar chart to visualize the distribution of price ranges among the restaurants.
data.hist(column = "Price range")
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Counts")
plt.show()
```

```python
# Questions 2 - Calculate the percentage of restaurants in each price range category.
total_rest = len(data)
cnt = data['Price range'].value_counts()
price_range = round(cnt / total_rest * 100, 2)
pr = pd.merge(cnt, price_range, on = "Price range")
print("The percentage of restaurants in price range are :", pr)
```

```python
# Pie Chart Visualization
plt.pie(price_range, labels = price_range.index, autopct = '%1.1f%%', colors = ['orange','skyblue','green','gold'])
plt.title("Price range percentage")
plt.show()
```
