```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea

print("Imported Successfully")
```

```python
# Loading data from dataset
data = pd.read_csv('Dataset.csv')

# columns present in dataset
print(data.columns)

data.head(3)
```

```python
                               # Question 1 - Identify the restaurants with the highest and lowest number of votes.

highest_rated = data.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending = False)
print("Highest Rated Resraurant is: ", highest_rated.idxmax())

lowest_rated = data.groupby('Restaurant Name')['Votes'].sum().sort_values()
print("Highest Rated Resraurant is: ", lowest_rated.idxmin())
```

```python
                     # Question 2 - Analyze if there is a correlation between the number of votes and the rating of a restaurant.

# Correlation between votes and rating
correlation = data['Votes'].corr(data['Aggregate rating'])
print(f"Correlation between Votes and Rating: {correlation:.2f}")
```

```python
# Visual representation
plt.figure(figsize = (8, 4), dpi = 100)
sea.heatmap(correlation, annot = True, fmt = ".2f", linewidth = .4)
plt.show()
```

```python
# Display the correlation coefficient
correlation = data['Votes'].corr(data['Aggregate rating'])
print("As per result correlation value = 0.31, this indicates there is a moderate positive correlation between votes and rating")

# Step 2: Visualize the relationship with a scatter plot
plt.figure(figsize=(10, 10))
plt.scatter(data['Votes'], data['Aggregate rating'], color='purple', alpha=0.6)
plt.title('Correlation between Votes and Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate rating')
plt.grid(True)
plt.show()
```
