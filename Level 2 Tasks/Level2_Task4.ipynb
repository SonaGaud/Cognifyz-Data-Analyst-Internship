```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea

print("Imported Successfully")
```

```python
# Loading data from dataset
data = pd.read_csv('Dataset.csv')

# columns present in dataset
print(data.columns)
```

```python
                                # Question 1 - Identify if there are any restaurant chains present in the dataset.

chains = data['Restaurant Name'].value_counts()
Restaurant_chains = chains[chains > 1]
print("The restaurants that have multiple branches are :\n", Restaurant_chains.head(10))
```

```python
# Plot bar chart
# showing top 5 restaurant chains
top_5_chains = Restaurant_chains.head(5)  

# Created the bar plot
plt.figure(figsize=(10, 3))
sea.barplot(x=top_5_chains.values, y=top_5_chains.index, palette='cool', edgecolor = "black")

# Labels and title
plt.xlabel("Number of Restaurants")
plt.ylabel("Restaurant Chain")
plt.title("Top 5 Most Common Restaurant Chains")

# Showing the plot
plt.show()
```

```python
# Question 2 - Analyze the ratings and popularity of different restaurant chains

# Average rating per restaurant chain
chain_ratings = data.groupby('Restaurant Name')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print("Top 10 Highest-Rated Restaurant Chains:\n", chain_ratings)

# Plot bar chart
plt.figure(figsize=(10,5))
sea.barplot(x=chain_ratings.values, y=chain_ratings.index, palette='cool')
plt.xlabel("Average Rating")
plt.ylabel("Restaurant Chain")
plt.title("Top 10 Highest-Rated Restaurant Chains")
plt.show()
```
